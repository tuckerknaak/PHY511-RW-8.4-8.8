```python
#Imports
import matplotlib.pyplot as plt
import numpy as np


#lambda = 500nm +- lambda_{f sr}
wavelength_range = np.linspace(250,750,100)


#Computing transmittance values
transmittance = []
d = 5*10**6  #mirror separation [nm]
F = 89.75    #coefficient of finesse

for wavelength in wavelength_range:
    t = 1/(1+F*(np.sin(2*np.pi*d/wavelength)**2))
    transmittance.append(t)

    
#Plotting the figure
plt.figure(figsize=(8,5),dpi=100)
plt.plot(wavelength_range, transmittance, linestyle='solid', color='red')
plt.grid(True, alpha=0.4, linestyle='dashed', color='darkgray')
plt.xticks([250,350,450,550,650,750])
plt.xlabel('Wavelength [nm]')
plt.ylabel('Transmittance')
plt.title('Transmittance as Function of Wavelength')
plt.show()
```
